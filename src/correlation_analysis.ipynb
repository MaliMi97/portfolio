```python
coin_list = ['bitcoin', 'ethereum', 'solana', 'polkadot', 'the-sandbox', 'decentraland']
c = CoinGeckoAPI()
end = datetime.now()
start = datetime(end.year-1, end.month, end.day)
```

```python
for coin in coin_list:
    globals()[coin+"_d"] = c.get_price_cap_volume(start, end, coin=coin, period='day')
    globals()[coin+"_w"] = c.get_price_cap_volume(start, end, coin=coin, period='week')
daily, weekly = pd.DataFrame(), pd.DataFrame()
for coin in coin_list:
    daily[coin+"_d"] = globals()[coin+"_d"]['price']
    weekly[coin+"_w"] = globals()[coin+"_w"]['price']
```
